```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = ['A','B','C','D','E'] #categories for the x-axis
values = [5,7,3,8,4] # corresponding values for the y-axis
width = [0.2,0.3,0.4,0.6,0.8] # width of the bars
colors = ['grey','orange','green','red','skyblue']

#plt.bars(x,y,color)
plt.bar(categories, values, width=width, color=colors, edgecolor='black')
plt.xlabel('student') # Label the x-axis
plt.ylabel('stats') # Label the y-axis
plt.title('mid term marks') # Title of the graph

plt.grid(axis='y', linestyle='--', alpha=0.7)

# edgecolor='black'
# Displaying the graph
plt.show()
```

**Line Graph**

```python
# Data for the line plot
x = [0,1,2,3,4,5] # x-axis values
y = [0,1,4,9,16,25] # y-axis values

# Plotting the line graph
plt.plot(x, y, color="red", linestyle='-', marker='o', label='y = x^2')

# Adding labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Example')

plt.legend()
plt.show()
```

**Scatter Plot**

```python
# Data for the scatter plot
x = [5,7,8,7,2,17,2,9,4,11] # x-axis values
y = [99,86,87,88,100,86,103,82,94,78] # y-axis values

# Creating the scatter plot
plt.scatter(x, y, color="red", alpha=0.7, edgecolor='black', label='Data points')

# Adding labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot Example')

plt.legend()

#Displaying the grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Customising Scatter Plot**

```python
# Data for the scatter plot
x = [5,7,8,7,2,17,2,9,4,11] # x-axis values
y = [99,86,87,88,100,86,103,82,94,78] # y-axis values
sizes = [20,50,100,200,500,100,60,90,30,120] # sizes of markers

# Creating the scatter plot
plt.scatter(x, y, color="red", s=sizes, alpha=0.7, edgecolor='black', label='Data points')

# Adding labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot Example')

plt.legend()

#Displaying the grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

**Histogram**

```python
#if wants to fix the random state
np.random.seed(30)
#Generate data
data = np.random.randn(1000)
#Create histogram
plt.hist(data, bins=20, color='red', alpha=0.7, edgecolor='black', label='Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Pie Chart**

```python
#Data
sizes = [40,30,20,10]
labels = ['Group 1','Group 2','Group 3','Group 4']
colors = ['grey','orange','green','red']
#Plot pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=(0,0,0,0.4))
plt.title('Pie Chart Example')
plt.show()
```

```python
w= 0.2
x=["C1","C2","C3","C4"]
boys=[100,40,60,50]
girls=[200,70,30,70]
bar1=np.arange(len(x))
plt.bar(bar1,boys,w,label="boys")
bar2=[i+w for i in bar1]
plt.bar(bar2,girls,w,label="girls")
plt.xticks(bar1+w/2,x)
plt.show()
```
